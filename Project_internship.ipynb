```python
import pandas as pd

df = pd.read_csv("HR-Employee-Attrition-All (1).csv")

df.head()
```

```python
df.isnull().sum()
df = df.drop_duplicates()

df.head()
```

```python
df['Attrition'].value_counts()
```

```python
df['Attrition'].value_counts()
```

```python
import matplotlib.pyplot as plt

df[df['Attrition']=="Yes"]['Department'].value_counts().plot(kind='bar')

plt.title("Department Wise Employee Attrition")
plt.xlabel("Department")
plt.ylabel("Number of Employees")

plt.show()
```

```python
df.groupby('Attrition')['MonthlyIncome'].mean()
df.groupby('Attrition')['YearsAtCompany'].mean()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,8))

plt.subplot(2,2,1)
df['Attrition'].value_counts().plot(kind='bar')
plt.title("Attrition Count")


plt.subplot(2,2,2)
df['Attrition'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%'
)
plt.title("Attrition Percentage")


plt.subplot(2,2,3)
df[df['Attrition']=="Yes"]['Department'].value_counts().plot(kind='bar')
plt.title("Department Attrition")


plt.subplot(2,2,4)
df.groupby('Attrition')['MonthlyIncome'].mean().plot(kind='bar')
plt.title("Income vs Attrition")


plt.tight_layout()
plt.show()
```
